```python
import sys
sys.path.append('../src')
from random import sample, seed

from agents import Man, Woman
from deferred_acceptance import deferred_acceptance
from all_rotations import all_rotations
from join_irreducible import join_irreducible_matching
from all_stable_matching import all_stable_matching
from median_stable_matching import median_stable_matching
```

```python
seed(0)

n = 80
men_list = [Man(i) for i in range(n)]
women_list = [Woman(i) for i in range(n)]

for man in men_list:
    man.set_pref(sample(women_list, k=len(women_list)))
for woman in women_list:
    woman.set_pref(sample(men_list, k=len(men_list)))

man_optimal_matching = deferred_acceptance(men_list, women_list, True)
print(man_optimal_matching)
woman_optimal_matching = deferred_acceptance(men_list, women_list, False)
print(woman_optimal_matching)

print(man_optimal_matching == woman_optimal_matching)

rotations = all_rotations(men_list, women_list)
print(len(rotations))
for r in rotations:
    print(r)
```

```python
join_irreducible_matchings = []
for rotation in rotations:
    join_irreducible_matchings.append(join_irreducible_matching(rotation, men_list, women_list))

len(join_irreducible_matchings)
for matching in join_irreducible_matchings:
    print(matching)
```

```python
all_stable_matchings = all_stable_matching(men_list, women_list)
print(len(all_stable_matchings))
for matching in all_stable_matchings:
    print(matching)
```

```python
def is_stable(matching, men_list, women_list):
    for m in men_list:
        for w in women_list:
            if (m.rank(w) < m.rank(matching[m])) and (w.rank(m) < w.rank(matching[w])):
                return False
    return True

def is_all_stable(matchings, men_list, women_list):
    for matching in matchings:
        if not is_stable(matching, men_list, women_list):
            return False
    return True

def all_distinct(matchings):
    for i, matching in enumerate(matchings):
        for j, matching2 in enumerate(matchings):
            if i == j:
                continue
            if matching == matching2:
                return False
    return True

print("全マッチングが安定である：", is_all_stable(all_stable_matchings, men_list, women_list))
print("全マッチングが全て異なる：", all_distinct(all_stable_matchings))
```

```python
median = median_stable_matching(men_list, women_list, all_stable_matchings, len(all_stable_matchings)//2)
print(is_stable(median, men_list, women_list))
print(median)
```
